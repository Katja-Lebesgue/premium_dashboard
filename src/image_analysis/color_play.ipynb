```python
def get_repreesentatives_of_n_subdivision_of_color_space(n: int):
    red_points = np.linspace(0, 1, n)
    green_points = np.linspace(0, 1, n)
    blue_points = np.linspace(0, 1, n)

    return list(product(red_points, green_points, blue_points))

def plot_color_palette(colors: list):
    fig = plt.figure(figsize=(20, 10))
    for i, color in enumerate(colors):
        fig.add_subplot(len(colors)//10+1, 10, i+1)
        plt.imshow([[color]])
        plt.axis('off')
```

```python
red = Color('red')
pink = Color('pink')
rainbow = list(red.range_to(pink, 30))
fig = plt.figure(figsize=(20, 10)))
fig = plt.figure(figsize=(20, 10))
for i, color in enumerate(rainbow):
    fig.add_subplot(1, len(rainbow), i+1)
    color.set_saturation(1)
    color.set_luminance(0.5)
    plt.imshow([[color.rgb]])
```
